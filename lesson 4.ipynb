__5*.__ Написать на Python программу с реализацией одного из изученных алгоритмов решения СЛАУ.

```python
import numpy as np
```

```python
def kramer_method (a, b):
    x = list()
    n, m = a.shape
    if n!=m:
        print('Матрица не квадратная, решение методом Крамера невозможно')
        return
    det_a = np.linalg.det(a)
    if det_a == 0:
        print('Определитель равен нулю, система вырождена')
        return
    for i in range(n):
        c = a.copy()
        c[:, i] = b
        x.append(np.linalg.det(c)/det_a)
    return np.array(x)
```

```python
a = np.array([[1,-2],[3,-4]])
b=np.array([1,7])
print(f'Матрица A:\n{a}')
print(f'Свободный член b:\n{b}')
print(f'Решение x:\n{kramer_method(a,b)}')
```

```python
a = np.array([[2,-1,5],[1,1,-3],[2,4,1]])
b=np.array([10,-2,1])
print(f'Матрица A:\n{a}')
print(f'Свободный член b:\n{b}')
print(f'Решение x:\n{kramer_method(a,b)}')
```
